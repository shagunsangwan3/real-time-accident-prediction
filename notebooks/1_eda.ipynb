```python
import pandas as pd 

df = pd.read_csv("../data/raw/US_Accidents_March23.csv")
df.head()
```

```python
df.columns
```

```python
selected_cols=[
    "Start_Time",
    "Start_Lat",
    "Start_Lng",
    "Severity",
    "Weather_Condition",
    "Visibility(mi)",
    "Wind_Speed(mph)",
    "Temperature(F)",
    "Precipitation(in)"
]

df = df[selected_cols]
df.head()
```

```python
df.isnull().sum()
```

```python
num_cols = [
     "Visibility(mi)",
    "Wind_Speed(mph)",
    "Temperature(F)",
    "Precipitation(in)"
]

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df["Weather_Condition"] = df["Weather_Condition"].fillna("Unknown")
```

```python
df["Weather_Condition"] = df["Weather_Condition"].astype("category").cat.codes
```

```python
df.isnull().sum()
```

```python
df['Severity'].value_counts()
```

```python
df.describe()
```

```python
# Convert Start_time to datetime--->>>>
df["Start_Time"] = pd.to_datetime(df["Start_Time"], errors="coerce")
```

```python
# Extract time-Based Features---->>>

df["hour"] = df["Start_Time"].dt.hour
df["day"] = df["Start_Time"].dt.day
df["month"] = df["Start_Time"].dt.month
df["weekday"] = df["Start_Time"].dt.weekday
```

```python
df["is_weekend"] = df["weekday"].isin([5,6]).astype(int)
df["is_night"] = df["hour"].apply(lambda x: 1 if (x >= 19 or x <= 6) else 0)
```

```python
# Create a location_based risk feature---->>>
df['lat_bin'] = df['Start_Lat'].round(2)
df['lng_bin'] = df['Start_Lng'].round(2)
```

```python
# Create Historical Accidents

location_risk = (
    df.groupby(['lat_bin','lng_bin'])
                .size()
                .reset_index(name="accident_count")
)

df = df.merge(location_risk, on=['lat_bin','lng_bin'],how='left')
```

```python
# Select features for Modeling(Independent or Dependent Features)--->>>>>

feature_cols = [
    "hour",
    "is_night",
    "is_weekend",
    "lat_bin",
    "lng_bin",
    "Weather_Condition",
    "Visibility(mi)",
    "Wind_Speed(mph)",
    "Temperature(F)",
    "Precipitation(in)",
    "accident_count"
]

X = df[feature_cols]
y = df['Severity']
```

```python
# Sort Data by time 
df = df.sort_values("Start_Time").reset_index(drop=True)
```

```python
# Keep only 300k rows (you can try 200k if needed as data is 6 gib) --->>>>>
df_small = df.iloc[:300_000].copy()
```

```python
X_small = df_small[feature_cols]
y_small = df_small["Severity"]
```

```python
# Scale Features

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X_small)
```

```python
# Define Sequence Length (Predict risk on base of past Data)---->>>>

SEQ_LEN = 10 # YOU CAN LATER TRY 20,30
```

```python
# Create Sequences--->>>

import numpy as np

def create_sequence(X_small,y_small,seq_len):
    X_seq = []
    y_seq = []

    for i in range(len(X_small) - seq_len):
         X_seq.append(X_small[i:i+seq_len])
         y_seq.append(y_small.iloc[i+seq_len])

    return np.array(X_seq), np.array(y_seq)
```

```python
X_seq,y_seq = create_sequence(X_scaled,y_small,SEQ_LEN)

print(X_seq.shape)
print(y_seq.shape)
```

```python
# Train-test Split--->>>

split_idx = int(0.8*len(X_seq))

X_train,X_test = X_seq[:split_idx], X_seq[split_idx:]
y_train,y_test = y_seq[:split_idx], y_seq[split_idx:]
```

```python
# Importing Deep Learning Libraries---->>>
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Build the LSTM architecture ----->>>>>

model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.3),

    LSTM(32),
    Dropout(0.3),

    Dense(1, activation="linear") # Predicting serverity (1-4)
    
])

model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

model.summary()
```

```python
# Train/test split 
early_stop = EarlyStopping(
    monitor = "val_loss",
    patience=3,
    restore_best_weights=True
)
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_test,y_test),
    epochs=10,
    batch_size=128,
    callbacks=[early_stop],
    verbose=1
)
```

```python
loss, mae = model.evaluate(X_test, y_test, verbose=0)
print(f"Test MAE: {mae:.3f}")
```
